```python
def iterate(k,root=''):
    
    global total_data
    children = [a for a in k.keys() if (len(a)>12) and a!='Reference_Guid']
    parent = [root]*len(children)

    try:
        reference = "-".join(k['Reference_Guid'])
    except:
        reference = ''
    cite = [reference]*len(children)
    
    total_data = total_data.append(pd.DataFrame({'child':children,'parent':parent,'reference':cite}))
    for i in children:
        iterate(k[i],i)
```
